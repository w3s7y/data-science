```python
import pandas

pt = pandas.read_csv('work/periodic-table/periodic_table.csv')
pt
```

```python
print(f"Mean melting point of all known elements = {pt.MeltingPoint.mean()}K")
print(f"Mean boiling point of all known elements = {pt.BoilingPoint.mean()}K")
```

```python
# Lets plot boiling points Kelvin (Y axis) as a function of atomic mass (X axis)
pt.plot(x=pt.AtomicMass.name, y=pt.BoilingPoint.name)
```

```python
# Lets find all the deets for carbon 
pt.loc[5]
```

```python
# Lets do a few visuals with the halides
halides = pt.loc[pt['Group'] == 17]
halides.plot(x=halides.Element.name, y=halides.MeltingPoint.name)
```

```python
# boxplot of first ionization energy for halide group elements
halides.boxplot([halides.FirstIonization.name])
```
